```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
def activation_func(yin):
    if yin >= 1:
        return 1
    else:
        return -1
```

```python
def plotting_points(a1,a2,b1,b2,r):
    plt.scatter(x=a1,y=a2,marker="*", s=300, c= "green")
    plt.scatter(x=b1,y=b2,marker='*', s=300, c="red")
    plt.xlabel("X1")
    plt.ylabel("X2") 
    if r==str(1):
        plt.title('OR GATE')
    elif r==str(2):
        plt.title('AND GATE')
    else:
        plt.title('AND NOT GATE')

    plt.show()
```

```python
#Input data
n = int(input("Enter number of input vectors: "))
x = []
y = []
for i in range(0,n):
    raw_str1 = str(input("Enter values for vector " + str(i+1) + ": "))
    input_vector = raw_str1.split(' ')
    ip_list = []
    for ele in input_vector:
        ip_list.append(float(ele))
    print(ip_list)
    np_list = np.array(ip_list, dtype=np.float64)
    x.append(np_list)
    curr_target = str(input("Enter the target for vector " + str(i+1) + ": "))
    y.append(curr_target)
x1,x2 = organising_inputs(x)
a1,a2,b1,b2 = separating_inputs(x,y)
```

```python
def plotting_line(a1,a2,b1,b2,p1,p2):
    plt.scatter(x=a1,y=a2,marker="*", s=300, c= "green")
    plt.scatter(x=b1,y=b2,marker='*', s=300, c="red")
    plt.plot(p1, p2, 'go-', label='sepration line', linewidth=2)
    plt.xlabel("X1")
    plt.ylabel("X2") 
    if r==str(1):
        plt.title('OR GATE')
    elif r==str(2):
        plt.title('AND GATE')
    else:
        plt.title('AND NOT GATE')

    plt.show()
```

```python
#Input coordinates of line that can seprarate the points linearly
p1 = [int(item) for item in input("Enter coordinate 1 of line : ").split()]
p2 = [int(item) for item in input("Enter coordinate 1 of line : ").split()]
```

```python
#calculating the slope and intercept using the points entered
m = (p2[1]-p1[1])/(p2[0]-p1[0])
c = p1[1] - m*p1[0]
```

```python
# Assuming weights and bias of the gate
w1 = 1
w2 = -1*(m/w1)
b = -1*(c*w2)
```
